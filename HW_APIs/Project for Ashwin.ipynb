```python

for st in States:
    query = "ELEC.GEN.ALL-" + st + "-99.A"
    # Build query URL
    query_url = url + "api_key=" + api_key + "&series_id=" + query
    print(query_url)
    articles = requests.get(query_url).json()
    
    good_columns=["Year_" + st, st]
    #print(articles["series"][0]["data"])
    d=articles["series"][0]["data"]
    
    dat2=pd.DataFrame(data=d,columns=good_columns)
    dat2=dat2.drop(columns=["Year_" + st])
    main_df= pd.concat([main_df,dat2],axis=1)

main_df=main_df.set_index('Year')
main_df        
    #for article in articles["data"]:
        #print(article)    
```
