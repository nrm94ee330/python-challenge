```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
print(purchase_data.head())
purchase_data.describe()
```

```python
unique_player_count=purchase_data["SN"].nunique()
print(f"Total Unique Players are: {unique_player_count}")
```

```python
unique_item_count=purchase_data["Item ID"].nunique()
#print(f"Total Unique Items are: {unique_item_count}")

Average_price_val=purchase_data["Price"].mean()
Average_price=f"${Average_price_val:2.3}"

Num_purchases=purchase_data["Purchase ID"].nunique()
#print(Num_purchases)

Tot_revenue=f"${(purchase_data['Price']*1).sum():,}"
#print(Tot_revenue)

df = pd.DataFrame({'Number of Unique Items': [unique_item_count], 'Average Price': [Average_price], 'Number of Purchases': [Num_purchases], 'Total Revenue':[Tot_revenue]}) 
df
```

```python
purchase_data_user_grp=purchase_data.groupby("SN")
purchase_age_data=purchase_data_user_grp["Age"].mean()
print(purchase_age_data.head())
purchase_price_data=purchase_data_user_grp["Price"].sum()
print(purchase_price_data)
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39,100]

# Create the names for the four bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]
purchase_age_data1 = pd.DataFrame({"Age": purchase_age_data,"Price":purchase_price_data})
                               

purchase_age_data1["Age Bins"]=pd.cut(purchase_age_data1["Age"], bins, labels=group_labels)

purchase_age_data1.head()
```

```python
purchase_age_data2 = purchase_age_data1.groupby("Age Bins")

# Find how many rows fall into each bin
age_count=purchase_age_data2["Age"].count()
total_age=purchase_age_data1["Age"].count()
avg_age_bin=round(purchase_age_data2["Age"].count()/total_age*100,2)

df_age = pd.DataFrame({          "Total Count": age_count,
                                    "Percentage of Players":avg_age_bin
                                   })
df_age
```

```python
purchase_data_user_grp=purchase_data.groupby("SN")
purchase_user_count=purchase_data_user_grp["Purchase ID"].count()
#print(purchase_user_count.head())
purchase_user_avg=purchase_data_user_grp["Price"].mean()
#print(purchase_user_avg.head())
purchase_user_total=purchase_data_user_grp["Price"].sum()
#print(purchase_user_total.head())

df_user2 = pd.DataFrame({          "Purchase Count": purchase_user_count
                                    ,"Average Purchase Price": purchase_user_avg
                                    ,"Total Purchase Value":purchase_user_total
                                   })
df_user2.sort_values("Purchase Count", ascending=False)
```

```python
purchase_data_user_grp=purchase_data.groupby("Item ID")
purchase_item_count=purchase_data_user_grp["Item ID"].count()
#print(purchase_user_count.head())
purchase_item_avg=purchase_data_user_grp["Price"].mean()
#print(purchase_user_avg.head())
purchase_item_total=purchase_data_user_grp["Price"].sum()
#print(purchase_user_total.head())

df_item2 = pd.DataFrame({         "Item Name": purchase_data_user_grp["Item Name"][2]
                                 ,"Purchase Count": purchase_item_count
                                    ,"Average Purchase Price": purchase_item_avg
                                    ,"Total Purchase Value":purchase_item_total
                                   })
df_item2.sort_values("Purchase Count", ascending=False)
df_item2.head()
```
